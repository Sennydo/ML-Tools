```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
classifier("I am really excited to watch a show tonight")
```

```python
classifyMultiple = pipeline("sentiment-analysis")
classifyMultiple(["I really want to go to watch a movie", "I didn't like that movie at all"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier("This is a course on political science, we'll explore various influences",
           candidate_labels=["education", "entertainment", "business", "politics"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier(["I went to watch a movie", "I read a new publication in Science Journal"],
           candidate_labels=["education", "entertainment", "business"])
```

Text Generation with HuggingFace

```python
from transformers import pipeline

txt = pipeline("text-generation")
txt("In order to write a book, I have to do")
```

```python
#Using Specific Models

from transformers import pipeline

generate = pipeline("text-generation", model="distilgpt2")
generate("After I woke up today I",
                    max_length = 50,
                    num_return_sequences=3)
```

Extracting Strings from the sequence

```python
from transformers import pipeline

newText = pipeline("text-generation", model="distilgpt2")
resultText = newText("After coming back from my baseball practice I went to take a shower",
        max_length=30,
        num_return_sequences=2)

#resultText is a list with dictionaries in it
storeText = []
for i in resultText:
  storeText.append(i['generated_text'])

for i in storeText:
  print(i)
```

## Mask Filling

```python
from transformers import pipeline

masking = pipeline("fill-mask")
masking("I went to <mask> to play basketball today", top_k=4)
```

Different Mask identifier

```python
from transformers import pipeline

masker = pipeline("fill-mask", model="bert-base-cased")
masker("I went to [MASK] and studied a lot", top_k=3)
```

## Named Recognition

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("George is a cool kid who likes to play baseball in California")
```

```python
from transformers import pipeline

answers = pipeline("question-answering")
answers(question="What does George live?", context="George is a student who is from a place called San Jose")
```
